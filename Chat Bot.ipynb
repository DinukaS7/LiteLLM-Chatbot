```python
from litellm import completion
import os
import tkinter as tk


os.environ["OPENAI_API_KEY"] = "API KEY"


def get_response():
    user_input = entry.get("1.0", tk.END).strip()
    if not user_input:
        output_box.delete("1.0", tk.END)
        output_box.insert(tk.END, "Please type a question first.")
        return

    try:
        messages = [{"content": user_input, "role": "user"}]
        response = completion(model="openai/gpt-4o", messages=messages)
        answer = response['choices'][0]['message']['content']
        output_box.delete("1.0", tk.END)
        output_box.insert(tk.END, answer)
    except Exception as e:
        output_box.delete("1.0", tk.END)
        output_box.insert(tk.END, f"Error: {str(e)}")


root = tk.Tk()
root.title("LiteLLM Chat")


tk.Label(root, text="Type your question:", font=("Segoe UI", 11, "bold")).pack(pady=5)

entry = tk.Text(root, height=4, width=60, font=("Segoe UI", 10))
entry.pack(padx=10, pady=5)

tk.Button(root, text="Get Answer", command=get_response, bg="#0078D7", fg="white", font=("Segoe UI", 10, "bold")).pack(pady=5)

tk.Label(root, text="Response:", font=("Segoe UI", 11, "bold")).pack(pady=5)
output_box = tk.Text(root, height=10, width=60, wrap=tk.WORD, font=("Segoe UI", 10))
output_box.pack(padx=10, pady=5)


root.mainloop()
```
